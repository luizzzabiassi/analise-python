````python
import pandas as pd
import plotly.express as px
````

````python
df_principal = pd.read_excel('/content/tabela-acoes.xlsx', sheet_name='Principal')
df_principal
````

````python
df_total_acoes = pd.read_excel('/content/tabela-acoes.xlsx', sheet_name='Total_de_acoes')
df_total_acoes
````

````python
df_chatgpt = pd.read_excel('/content/tabela-acoes.xlsx', sheet_name='ChatGPT')
df_chatgpt
````

````python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
````

**Aula 3 - Criação de colunas e manipulação dos dados:**

````python
df_principal = df_principal.rename(columns={'Último (R$)':'ValorFinal', 'Var. Dia (%)':'VarDiaPct'}).copy()
df_principal
````

````python
df_principal['VarPct'] = df_principal['VarDiaPct'] / 100
df_principal['ValorInicial'] = df_principal['ValorFinal'] / (df_principal['VarPct'] + 1)
df_principal
````

````python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
````

````python
df_principal = df_principal.drop(columns=['Código'])
df_principal
````

````python
df_principal['VariacaoRS'] = (df_principal['ValorFinal'] - df_principal['ValorInicial']) * df_principal['Qtde. Teórica']
df_principal
````

````python
pd.options.display.float_format = '{:.2f}'.format
````

````python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
````

````python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'QtdTeorica'}).copy()
df_principal
````

````python
df_principal['Resultado'] = df_principal['VariacaoRS'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
````

````python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
````

````python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da empresa'])
df_principal = df_principal.drop_duplicates()
df_principal
````

````python
df_principal = df_principal.rename(columns={'Idade (anos)':'Idade'}).copy()
df_principal
````

````python
df_principal['CategoriaIdade'] = df_principal['Idade'].apply(lambda x: 'Mais de 100 anos' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
df_principal
````

````python
# Calculando o maior valor
maior = df_principal['VariacaoRS'].max()

# Calculando o menor valor
menor = df_principal['VariacaoRS'].min()

# Calculando a média
media = df_principal['VariacaoRS'].mean()

# Calculando a média de quem subiu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['VariacaoRS'].mean()

# Calculando a média de quem desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['VariacaoRS'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
````

````python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
````

````python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['VariacaoRS'].sum().reset_index()
df_analise_segmento
````

````python
fig = px.pie(df_analise_segmento, names='Segmento', values='VariacaoRS', title='Variação em Reais por Segmento')
fig.show()
````

````python
df_analise_saldo = df_principal.groupby('Resultado')['VariacaoRS'].sum().reset_index()
df_analise_saldo
````

````python
fig = px.bar(df_analise_saldo, x='Resultado', y='VariacaoRS', text='VariacaoRS', title='Variação em Reais por Resultado')
fig.update_traces(texttemplate='R$ %{text:,.2f}')
fig.show()
````

````python
df_analise_idade = df_principal.groupby('CategoriaIdade')['VariacaoRS'].sum().reset_index()
df_analise_idade
````

````python
fig = px.bar(df_analise_idade, x='CategoriaIdade', y='VariacaoRS', text='VariacaoRS', title='Variação em Reais por Faixa Etária')
fig.update_traces(texttemplate='R$ %{text:,.2f}')
fig.show()
````


O método texttemplate em Plotly é utilizado para formatar o texto exibido em diferentes elementos do gráfico, como pontos de dados, rótulos de eixos, entre outros. Você pode usá-lo para personalizar o texto exibido em um gráfico de barras.

Aqui está um exemplo de como usar texttemplate em um gráfico de barras:

```
import plotly.graph_objects as go

# Dados de exemplo
x = ['A', 'B', 'C', 'D']
y = [10000, 20000, 30000, 40000]

# Criar o gráfico de barras
fig = go.Figure([go.Bar(x=x, y=y, text=y, textposition='auto', marker_color='skyblue')])

# Atualizar o texto exibido nas barras com um template personalizado
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')  # O .2s mostra os dois primeiros dígitos

# Exibir o gráfico
fig.show()
```

Neste exemplo, texttemplate='%{text:.2s}' formata o texto exibido nas barras para mostrar apenas os dois primeiros dígitos do valor. Você pode ajustar o template conforme necessário para atender aos seus requisitos de formatação.

O parâmetro textposition='outside' define a posição do texto nas barras para ser fora delas. Isso garante que o texto não se sobreponha às barras.

Você pode personalizar ainda mais o texttemplate para atender às suas necessidades específicas de formatação, incluindo adicionar unidades, prefixos, sufixos, etc.
